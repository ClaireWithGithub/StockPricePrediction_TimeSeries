```python
import pandas as pd
import numpy as np
from matplotlib.pylab import rcParams
import matplotlib.pyplot as plt
from pyramid.arima import auto_arima
from sklearn.model_selection import train_test_split


pd.set_option('max_columns', None)
%matplotlib inline
```

```python
#setting figure size
rcParams['figure.figsize'] = 20,10
```

```python
df = pd.read_csv("prices-split-adjusted.csv")
```

```python
df.head()
```

## set the "date" column as index

```python
df['date'] = pd.to_datetime(df.date,format='%Y-%m-%d')
df.index = df['date']
```

```python
df.head()
```

```python
symbol_names = df.symbol.unique()
len(symbol_names)
```

## Choose GOOGLE as our target company

```python
df.loc[df["symbol"] == "MSFT"].shape
```

```python
googledf = df.loc[df["symbol"] == "MSFT"]
googledf.head()
```

## Plot the historical Stock price of GOOGLE

```python
plt.figure(figsize=(16,8))
plt.plot(googledf['close'], label='Close Price history',)
```

```python
train, valid = train_test_split(googledf, test_size = 0.25, shuffle=False)
```

```python
train.head()
```

```python
# data = df.loc[df["symbol"] == "A"].sort_index(ascending=True, axis=0)
# train, valid = train_test_split(data, test_size = 0.3, shuffle=False)
# train = data[:1175]
# valid = data[1175:]

training = train['close']
validation = valid['close']

# model_step = auto_arima(training, start_p=1, start_q=1,
#                       #test='adf',       # use adftest to find optimal 'd'
#                       max_p=3, max_q=3, # maximum p and q
#                       m=12,             # frequency of series
#                       d=None,           # let model determine 'd'
#                       seasonal=True,   # Set Seasonality
#                       start_P=0, 
#                       D=1, 
#                       trace=True,
#                       error_action='ignore',  
#                       suppress_warnings=True, 
#                       stepwise=True)

model_step = auto_arima(training, start_p=1, start_q=1,
                      test='adf',       # use adftest to find optimal 'd'
                      max_p=3, max_q=3, # maximum p and q
                      m=12,             # frequency of series
                      d=1,           # let model determine 'd'
                      seasonal=True,   # Set Seasonality
                      start_P=0, 
                      D=1, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)
```

```python
model_step.fit(training)

forecast_step = model_step.predict(n_periods=len(validation))
forecast_step = pd.DataFrame(forecast_step,index = valid.index,columns=['Prediction'])
```

```python
print(model_step.summary())#3
```

```python
rms=np.sqrt(np.mean(np.power((np.array(valid['close'])-np.array(forecast_step['Prediction'])),2)))
rms
```

```python
plt.plot(train['close'])
plt.plot(valid['close'])
plt.plot(forecast_step['Prediction'])
```

```python
conint = model_step.predict(len(validation),return_conf_int=True)
```

```python
#plot

plt.plot(train['close'])
plt.plot(valid['close'])

plt.plot(forecast_step['Prediction'])

#plot confidence interval
plt.fill_between(valid.index, conint[1][:,0], conint[1][:,1], 
                 color='k', alpha=.08)

plt.xlabel('Time')
plt.ylabel('Close Price')
```
